```python
'''
Zip function from document:
Returns an iterator of tuples, where the i-th tuple contains the i-th element from each of the arguments sequences or iterables.
The iterator stops when the shortest input iterable is exhausted. With a simple iterable argument, it returns an iterator of 1-tuples.
With no arguments, it returns an empty iterator

==> This mean the length of input is decided by the shortest iterable rather than the longest
==> You still has a tuples even there is an 1 element in each of them
'''
```

```python
# Example: Zip fucntion 
prices = [10, 20, 439]
car_sizes =['small','medium', 'big']
colors=['blue', 'green']

```

```python
zip(prices, car_sizes)
```

```python

example1 = list(zip(prices, car_sizes))
example1
```

```python
# Example of shortest iterables when use zip fucntion
example2 = list(zip(prices, car_sizes, colors))
example2
```

```python
'''
To appear the longest iterables, use the intertools from zip_longest
import the zip_longest from itertools
'''
```

```python
from itertools import zip_longest
```

```python
# Compare:
list(zip(prices, car_sizes, colors))
```

```python
# Versus:
list(zip_longest(prices, car_sizes, colors))
```

```python
# To filling out the None value in zip_longest:
list(zip_longest(prices, car_sizes, colors, fillvalue= '?'))
```

```python
#Use case when use zip function:
'When there are two lists and want to use to sort the values in two lists but still want to reserve connection between two list'
'''
Step1: Combine use zip fucntion with sort
Step2: Unzip with the list has been sorted
'''
```

```python
letters = ['b', 'f', 'g', 'a']
num =[4,3,1,7]
```

```python
# Step1 : To sort both list:

data = sorted(zip(letters, num))
print(data)
```

```python
# Step2: Unzip with the list has been sorted out:
letters, num = zip(*data)
print(letters)
print(num)
```

```python
data2 = sorted(zip(num, letters))
print(data2)
```

```python
# Use case 2: use dictionary 
dict1= dict(zip(letters, num))
dict1
```
